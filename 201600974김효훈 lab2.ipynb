```python
from matplotlib import pyplot as plt
import math
import numpy as np

def f1(x):
    return x**2 

def f2(x):
    return x**3 + x**2 + x

def f3(x):
    return x * np.sin((x * x * x) + 3)

def f4(x):
    return np.cos(np.sqrt(e**x + 1)/2)    

def derivatative_solver(f,x):
    h=1e-10
    return  (f(x+h)-f(x))/h

print("%f"%derivatative_solver(f1,2))
print("%f"%derivatative_solver(f2,2))
print("%f"%derivatative_solver(f3,2))
print("%f"%derivatative_solver(f4,2))

x = np.arange(-10,10,0.01)

plt.legend(loc='upper center')
plt.subplot(221)
plt.plot(x,f1(x),label="f1(x)")
plt.plot(x,derivatative_solver(f1,x),label="f1(x)d")

plt.legend(loc='upper center')
plt.subplot(222)
plt.plot(x,f2(x),label="f2(x)")
plt.plot(x,derivatative_solver(f2,x),label="f2(x)d")

plt.legend(loc='upper center')
plt.subplot(223)
plt.plot(x,f3(x),label="f3(x)")
plt.plot(x,derivatative_solver(f3,x),label="f3(x)d")

plt.legend(loc='upper center')
plt.subplot(224)
plt.plot(x,f4(x),label="f4(x)")
plt.plot(x,derivatative_solver(f4,x),label="f4(x)d")

plt.legend(loc='upper center')
plt.yticks(range(-20,100,20))
plt.xticks(range(-10,10,1))


plt.show()
```
