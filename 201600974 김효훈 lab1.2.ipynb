```python
myList1 = ["Kim",16,22]
myList2 = ["Na",15,21]
dic = {}


for i in range(len(myList1)):
    if (type(myList1[i]) == type("")):
        if (myList1[i] not in dic):
            dic.setdefault(myList1[i])
            dic[myList1[i]] = [myList2[i+1],myList2[i+2]]
            
            
for i in range(len(myList2)):
    if (type(myList2[i]) == type("")):
        if(myList2[i] not in dic):
            dic.setdefault(myList2[i])
            dic[myList2[i]] = [myList1[i+1],myList1[i+2]]
                        
print(dic)
```
