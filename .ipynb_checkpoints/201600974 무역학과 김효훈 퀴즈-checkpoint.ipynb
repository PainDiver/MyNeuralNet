```python
#문제4-1 
import numpy as np

def relu(x): 
    return (x>0)*x 

def FowardPropagation(Input,hidden): #인풋을 가중치와 점곱해준 후 활성화함수 적용
    Xhidden = np.dot(hidden,Input)
    Ohidden = relu(Xhidden)
    return np.array(Ohidden)

Input = [[1.0],[0.5]]
winput_output = [[0.9,0.3],[0.2,0.8]]
FowardPropagation(Input,winput_output)
```

```python
#문제 4-2

target = [[1.85],[0.1]]

def Backpropagation(weight,Error): 
    weighted_Error = np.copy(weight)
    for i in range(len(weighted_Error)):    # 가중치의 행 수 만큼, 분모를 알맞게 나누는 과정
        weighted_Error[i] = np.divide(weighted_Error[i],sum(weighted_Error[i]))   #노드의 가중치를 더한값으로 나눔
    weighted_Error = np.transpose(weighted_Error) #나눈다음에 전치행렬을 해줌, 그래야 위 계산이 가능
    return np.dot(weighted_Error,Error)
    
output = FowardPropagation(Input,winput_output) #순전파 진행
Eoutput= np.abs(target- output)            #오차함수는 절대값으로
Ehidden = Backpropagation(winput_output,Eoutput)  #역전파 진행 후 오차값 구하기, 정규화 o
print(Ehidden)
```

```python
#문제 4-3
L =0.1

def derivative_relu(x):         #렐루함수의 미분
    return 2*((x>0)*x)


Input = [[1.0],[0.5]]
winput_output = [[0.9,0.3],[0.2,0.8]]
output = FowardPropagation(Input,winput_output)
#순전파 끝
Eoutput= np.abs(target- output)
Ehidden = Backpropagation(winput_output,Eoutput)
#역전파 끝


winput_output += L * numpy.dot( (Eoutput * derivative_relu(output)) , numpy.transpose(Input)  )
#가중치 업데이트

print(winput_output)
```
