```python
import csv

print('\n#1')

f = open("lab1.1.csv","w")

myList1 = ["apple","orange","banana"]
myList2 = ["basketball","Soccer","Game"]
myList3 = ["computer","fridge","airconditioner"]

myList1 += myList2+myList3

for com in range(len(myList1)):
    print('%d: %s' %(com,myList1[com]))
        
for i in range(len(myList1)):
    if(i+1!=len(myList1)):
        f.write(myList1[i]+",")
    else:
        f.write(myList1[i])
        
f.close()

f = open("lab1.1.csv","r")
rd = csv.reader(f)

print('\n#2')

for i in rd:
    for j in range(len(i)):
        print('%d: %s' %(j,i[j]))
    
f.close()
```
