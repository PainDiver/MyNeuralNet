```python
def vec_dot(vec1,vec2):
    assert (len(vec1) == len(vec2))
    v = [v1*v2 for v1,v2 in zip(vec1,vec2)]
    return sum(v)


def transpose(vec):
    row = len(vec)
    col = len(vec[0])
    
    v3 = []
    v4 = []
    for i in range(col):
        for j in range(row):
            v3 += [vec[j][i]]
        v4 += [v3]
        v3 =[]
       
    return v4   


v1 =[[1,5],
     [6,7]]

v2 =[[5,3],
     [7,2]]

def mul_mat(vec1:list,vec2:list):
    assert len(vec1[0])==len(vec2),"size not matched"
    multimes =len(vec1[0]) 
    v3 = transpose(vec2)
    v4 = []
    res = []
    for i in range(multimes):
        for j in range(multimes):
            res += [vec_dot(vec1[i],v3[j])]
        v4 += [res]
        res=[]
        
            
    return v4
  
v3 =[[1,5],
     [5,7]]

v4 =[[2,6],
     [8,6]]
        

v5 = mul_mat(v3,v4)
print(v5)
```
