```python
import numpy as np
from matplotlib import pyplot as plt

%pylab inline

from collections import Counter

import matplotlib

def make_simple_line_chart():
    
    years =[1950,1960,1970,1980,1990,2000,2010]
    gdp = [300.2,543.3,1075.9,2862.5,5979.6,10289.7,14958.3]
    
    plt.plot(years,gdp,color="green",marker="v",linestyle ="solid")
    
    plt.title("abstract GDP")
    
    plt.ylabel("Billions of $")
    plt.show()
    
make_simple_line_chart()
```

```python
def make_simple_bar_chart():
    from collections import Counter
    
    movies = ["Annie Hall","Ben-Hur","Casablanca","Gandhi","West Side Story"]
    num_oscars = [5,11,3,8,10]
    
    
    xs =[i+0.5 for i, _ in enumerate(movies)]
    
    plt.bar(xs, num_oscars)
    plt.ylabel("# of Academy Awards")
    plt.title("My Favorite Movies")
    
    plt.xticks([i+0.5 for i, _ in enumerate(movies)],movies)
    
    plt.show()

make_simple_bar_chart()
    
```

```python
def make_histogram():
    
    grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
    decile = lambda grade: grade
    histogram = Counter(decile(grade) for grade in grades)
    
    plt.bar([x-4 for x in histogram.keys()], histogram.values(),8)
    plt.axis([-5,105,0,5])
    plt.xticks([10*i for i in range(11)])
    plt.ylabel("# of Students")
    plt.title("Distribution of Exam 1 Grades")
    plt.show()

make_histogram()
```

```python
def make_several_line_charts():
    variance = [1,2,4,8,16,32,64,128,256]
    bias_squared = [256,128,64,32,16,8,4,2,1]
    total_error = [x+y for x,y in zip(variance,bias_squared)]
    
    xs = range(len(variance))
    
    plt.plot(xs,variance, 'g-', label='variance')
    plt.plot(xs,bias_squared, 'r-', label='bias^2')
    plt.plot(xs,total_error, 'b:', label='total error')
    
    plt.legend(loc=9)
    plt.xlabel("model complexity")
    plt.title("The Bias-Variance Tradeoff")
    plt.show()
    
make_several_line_charts()
```

```python
def make_scatter_plot():
    
    friends = [70,65,72,63,71,64,60,64,67]
    minutes = [175,170,205,120,220,130,105,145,190]
    labels = ['a','b','c','d','e','f','g','h','i']
    
    plt.scatter(friends,minutes)
    
    for label,friend_count, minutes_count in zip(labels,friends,minutes):
        plt.annotate(label,xy=(friend_count,minutes_count), xytext=(5,-5), textcoords='offset points')
    plt.title("Daily Minutes vs. Number of Friends")
    plt.xlabel("# of friends")
    plt.ylabel("daily minutes spent on the site")
    plt.show()
    
make_scatter_plot()
```

```python
import csv
f = open("data.csv","r")
rd = csv.reader(f)

xy =[]
for xpyp in rd:
      xy = np.append(xy,xpyp)

x =xy[::2]
y =xy[1::2]

xp=[]
yp=[]
for i in range(len(x)):
    xp.append(float(x[i]))
    yp.append(float(y[i]))
    
plt.bar(xp,yp)
plt.yticks(np.arange(0,100,20))
plt.xticks(np.arange(0,100,20))
plt.xlabel("x")
plt.ylabel("y")
plt.title("x,y position")

plt.show()
f.close()
```
