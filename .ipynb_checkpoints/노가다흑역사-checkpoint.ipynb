```python
import csv,pandas
import numpy as np
from matplotlib import pyplot as plt

def GetAverage(lis):
    su = 0
    for i in range(0,len(lis)):
        su += lis[i]
    su = su/len(lis)
    return su


f= open("iris.csv",'r')
rd=csv.reader(f)

inputdata = pandas.Series({
            "받침길이":[],
            "받침넓이":[],
            "꽃잎길이":[],
            "꽃잎넓이":[],
             "종류":[]
            })

seatosadata = pandas.Series({
            "받침길이":[],
            "받침넓이":[],
            "꽃잎길이":[],
            "꽃잎넓이":[],
            "종류" :[]
            })
                            
versicolordata = pandas.Series({
            "받침길이":[],
            "받침넓이":[],
            "꽃잎길이":[],
            "꽃잎넓이":[],
            "종류" :[]
            })

virginicadata = pandas.Series({
            "받침길이":[],
            "받침넓이":[],
            "꽃잎길이":[],
            "꽃잎넓이":[],
            "종류" :[]
            })

for i in rd:
    inputdata["받침길이"] += i[::5]
    inputdata["받침넓이"] += i[1::5]
    inputdata["꽃잎길이"] += i[2::5]
    inputdata["꽃잎넓이"] += i[3::5]
    inputdata["종류"] += i[4::5]

    
for i in range(0,150):
    inputdata["받침길이"][i] = float(inputdata["받침길이"][i])
    inputdata["받침넓이"][i] = float(inputdata["받침넓이"][i])
    inputdata["꽃잎길이"][i] = float(inputdata["꽃잎길이"][i])
    inputdata["꽃잎넓이"][i] = float(inputdata["꽃잎넓이"][i])
    
for i in range(0,5):
    seatosadata[i] = inputdata[i][0:50]
    versicolordata[i] = inputdata[i][50:100]
    virginicadata[i] = inputdata[i][100:150]

averageSL = [GetAverage(seatosadata["받침길이"]),GetAverage(versicolordata["받침길이"]),GetAverage(virginicadata["받침길이"])]
averageSW = [GetAverage(seatosadata["받침넓이"]),GetAverage(versicolordata["받침넓이"]),GetAverage(virginicadata["받침넓이"])]
averagePL = [GetAverage(seatosadata["꽃잎길이"]),GetAverage(versicolordata["꽃잎길이"]),GetAverage(virginicadata["꽃잎길이"])]
averagePW = [GetAverage(seatosadata["꽃잎넓이"]),GetAverage(versicolordata["꽃잎넓이"]),GetAverage(virginicadata["꽃잎넓이"])]

    
    
plt.subplot(221)
plt.scatter(inputdata["받침길이"],inputdata["꽃잎길이"])
plt.title("relation between sepal length and petal length")
plt.ylabel("petal length")
plt.xlabel("sepal length")
plt.yticks(range(0,14,2))
plt.xticks(range(0,14,2))
plt.show()


plt.subplot(222)
label =["SL","SW","PL","PW"]
gap1 = [2,7,12,17]
gap2 = [3,8,13,18]
gap3 = [4,9,14,19]

plt.bar(gap1,[averageSL[0],averageSW[0],averagePL[0],averagePW[0]])
plt.bar(gap2,[averageSL[1],averageSW[1],averagePL[1],averagePW[1]])
plt.bar(gap3,[averageSL[2],averageSW[2],averagePL[2],averagePW[2]])

plt.legend(['seatosa','versicolor','virginica'],loc='upper right')

plt.yticks(range(0,13,1))
plt.xticks(range(0,20))
plt.xticks([3,8,13,18],label)

plt.subplot(224)

label =["SL","PL"]
gap1 = [2,7]
gap2 = [3,8]
gap3 = [4,9]

plt.bar(gap1,[averageSL[0],averagePL[0]])
plt.bar(gap2,[averageSL[1],averagePL[1]])
plt.bar(gap3,[averageSL[2],averagePL[2]])

plt.legend(['seatosa','versicolor','virginica'],loc='upper right')

plt.yticks(range(0,13,1))
plt.xticks(range(0,20))
plt.xticks([3,8],label)

# SL과 PL간 비교 후 : 붓꽃의 꽃잎길이와 받침길이가 정비례함
# 이는 꽃의 받침의 길이로 꽃잎의 길이를 어느정도 예상 할 수 있게 한다.



traindata = pandas.Series({
            "받침길이":[],
            "받침넓이":[],
            "꽃잎길이":[],
            "꽃잎넓이":[],
            "종류" :[]
            })
testdata = pandas.Series({
            "받침길이":[],
            "받침넓이":[],
            "꽃잎길이":[],
            "꽃잎넓이":[],
            "종류" :[]
            })

for i in range(0,5):
    testdata[i]+= inputdata[i][0:12]
    testdata[i]+= inputdata[i][50:62]
    testdata[i]+= inputdata[i][100:112]
    
    traindata[i]+=inputdata[i][12:50]
    traindata[i]+=inputdata[i][62:100]
    traindata[i]+=inputdata[i][112:150]
    
testdata.to_csv("test_dataset.csv",sep=',',na_rep='NaN')
traindata.to_csv("train_dataset.csv",sep=',',na_rep='NaN')
```

```python
import csv,pandas
import numpy as np
from matplotlib import pyplot as plt

testdata = pandas.read_csv("test_dataset.csv",sep=',',delimiter =',',header=None,index_col=0,squeeze=True)
traindata = pandas.read_csv("train_dataset.csv",sep=',',delimiter=',',header=None,index_col=0,squeeze=True)

def GetAverage(lis):
    su = 0
    for i in range(0,len(lis)):
        su += lis[i]
    su = su/len(lis)
    return su


testseatosa = pandas.Series({
            "받침길이":[],
            "받침넓이":[],
            "꽃잎길이":[],
            "꽃잎넓이":[],
            "종류" :[]
            })

testversicolor = pandas.Series({
            "받침길이":[],
            "받침넓이":[],
            "꽃잎길이":[],
            "꽃잎넓이":[],
            "종류" :[]
            })

testvirginica = pandas.Series({
            "받침길이":[],
            "받침넓이":[],
            "꽃잎길이":[],
            "꽃잎넓이":[],
            "종류" :[]
            })

trainseatosa = pandas.Series({
            "받침길이":[],
            "받침넓이":[],
            "꽃잎길이":[],
            "꽃잎넓이":[],
            "종류" :[]
            })

trainversicolor = pandas.Series({
            "받침길이":[],
            "받침넓이":[],
            "꽃잎길이":[],
            "꽃잎넓이":[],
            "종류" :[]
            })

trainvirginica = pandas.Series({
            "받침길이":[],
            "받침넓이":[],
            "꽃잎길이":[],
            "꽃잎넓이":[],
            "종류" :[]
            })

for i in range(0,5):
    testdata[i]=testdata[i][1:len(testdata[i])-1].split(",")
    traindata[i]=traindata[i][1:len(traindata[i])-1].split(",")

print(len(testdata["받침길이"]))
print(len(traindata["받침길이"]))
    
for i in range(0,len(testdata["받침길이"])):
    testdata["받침길이"][i] = float(testdata["받침길이"][i])
    testdata["받침넓이"][i] = float(testdata["받침넓이"][i])
    testdata["꽃잎길이"][i] = float(testdata["꽃잎길이"][i])
    testdata["꽃잎넓이"][i] = float(testdata["꽃잎넓이"][i])
    
for i in range(0,len(traindata["받침길이"])):
    traindata["받침길이"][i] = float(traindata["받침길이"][i])
    traindata["받침넓이"][i] = float(traindata["받침넓이"][i])
    traindata["꽃잎길이"][i] = float(traindata["꽃잎길이"][i])
    traindata["꽃잎넓이"][i] = float(traindata["꽃잎넓이"][i])
    
for i in range(0,5):
    trainseatosa[i] = traindata[i+1][0:38]
    trainversicolor[i] = traindata[i+1][38:76]
    trainvirginica[i] = traindata[i+1][76:114]
    
    testseatosa[i] = testdata[i+1][0:12]
    testversicolor[i] = testdata[i+1][12:24]
    testvirginica[i] = testdata[i+1][24:36]



testaverageSL = [GetAverage(testseatosa["받침길이"]),GetAverage(testversicolor["받침길이"]),GetAverage(testvirginica["받침길이"])]
testaverageSW = [GetAverage(testseatosa["받침넓이"]),GetAverage(testversicolor["받침넓이"]),GetAverage(testvirginica["받침넓이"])]
testaveragePL = [GetAverage(testseatosa["꽃잎길이"]),GetAverage(testversicolor["꽃잎길이"]),GetAverage(testvirginica["꽃잎길이"])]
testaveragePW = [GetAverage(testseatosa["꽃잎넓이"]),GetAverage(testversicolor["꽃잎넓이"]),GetAverage(testvirginica["꽃잎넓이"])]

trainaverageSL = [GetAverage(trainseatosa["받침길이"]),GetAverage(trainversicolor["받침길이"]),GetAverage(trainvirginica["받침길이"])]
trainaverageSW = [GetAverage(trainseatosa["받침넓이"]),GetAverage(trainversicolor["받침넓이"]),GetAverage(trainvirginica["받침넓이"])]
trainaveragePL = [GetAverage(trainseatosa["꽃잎길이"]),GetAverage(trainversicolor["꽃잎길이"]),GetAverage(trainvirginica["꽃잎길이"])]
trainaveragePW = [GetAverage(trainseatosa["꽃잎넓이"]),GetAverage(trainversicolor["꽃잎넓이"]),GetAverage(trainvirginica["꽃잎넓이"])]

plt.subplot(221)
label =["SL","SW","PL","PW"]
gap1 = [2+5*i for i in range(0,4)]
gap2 = [3+5*i for i in range(0,4)]
gap3 = [4+5*i for i in range(0,4)]

plt.bar(gap1,[testaverageSL[0],testaverageSW[0],testaveragePL[0],testaveragePW[0]])
plt.bar(gap2,[testaverageSL[1],testaverageSW[1],testaveragePL[1],testaveragePW[1]])
plt.bar(gap3,[testaverageSL[2],testaverageSW[2],testaveragePL[2],testaveragePW[2]])


plt.legend(['seatosa','versicolor','virginica'],loc='upper right')
plt.yticks(range(0,13,1))
plt.xticks(range(0,20))
plt.xticks([3+5*i for i in range(0,4)],label)


plt.subplot(222)
label =["SL","SW","PL","PW"]
gap1 = [2+5*i for i in range(0,4)]
gap2 = [3+5*i for i in range(0,4)]
gap3 = [4+5*i for i in range(0,4)]

plt.bar(gap1,[trainaverageSL[0],trainaverageSW[0],trainaveragePL[0],trainaveragePW[0]])
plt.bar(gap2,[trainaverageSL[1],trainaverageSW[1],trainaveragePL[1],trainaveragePW[1]])
plt.bar(gap3,[trainaverageSL[2],trainaverageSW[2],trainaveragePL[2],trainaveragePW[2]])

plt.legend(['seatosa','versicolor','virginica'],loc='upper right')
plt.yticks(range(0,13,1))
plt.xticks(range(0,20))
plt.xticks([3+5*i for i in range(0,4)],label)

columns = {'test-setosa':[testaverageSL[0],testaverageSW[0],testaveragePL[0],testaveragePW[0]],
          'train-setosa':[trainaverageSL[0],trainaverageSW[0],trainaveragePL[0],trainaveragePW[0]],
           'test-verisoclor':[testaverageSL[1],testaverageSW[1],testaveragePL[1],testaveragePW[1]],
          'train-verisoclor':[trainaverageSL[1],trainaverageSW[1],trainaveragePL[1],trainaveragePW[1]],
           'test-virginica':[testaverageSL[2],testaverageSW[2],testaveragePL[2],testaveragePW[2]],
          'train-virginica':[trainaverageSL[2],trainaverageSW[2],trainaveragePL[2],trainaveragePW[2]]
          }

info = pandas.DataFrame(columns,["SL","SW","PL","PW"])
info


# 테스트 집단, 트레인 집단 모두 절대적인 길이나 넓이가 평균적으로 크게 차이는 나지 않지만, 두 집단 모두 꽃잎의 길이와  받침길이가 정비례한다는 것을 알 수 있다.
```
