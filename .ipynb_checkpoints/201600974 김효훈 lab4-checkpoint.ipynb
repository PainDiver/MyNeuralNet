```python
import pandas
import numpy as np
from matplotlib import pyplot as plt


irisdf= pandas.read_csv("iris.csv",names=['SL','SW','PL','PW','species'] ,sep=",")

irisdf
```

```python
sl = irisdf.iloc[:,0:1]
sw = irisdf.iloc[:,1:2]
pl = irisdf.iloc[:,2:3]
pw = irisdf.iloc[:,3:4]

plt.scatter(sl,pl)
plt.xlabel("sepal length")
plt.ylabel("petal length")
plt.title("petal legnth in proportion to sepal length")
plt.show()

#꽃받침 길이에 따른 꽃잎 길이 -> 꽃받침 길이를 보고 꽃잎 성장을 예측할 수 있게 됨.
#선형적인 결과가 대체적으로 많이 보임 -> 유효
```

```python
sg = irisdf.groupby("species")
sgm = sg.mean()

plt.subplot(221)
plt.bar([i for i in range(0,3)],sgm["SL"],width=0.1)
plt.bar([i+0.1 for i in range(0,3)],sgm["SW"],width=0.1)
plt.bar([i+0.2 for i in range(0,3)],sgm["PL"],width=0.1)
plt.bar([i+0.3 for i in range(0,3)],sgm["PW"],width=0.1)
plt.xticks([i+0.1 for i in range(0,3)],["setosa","versicolor","virginica"])
plt.title("mean by speicies")
plt.legend(["SL","SW","PL","PW"],fontsize =7)
plt.show()

plt.subplot(222)
plt.scatter(["setosa","versicolor","virginica"],sgm["SL"])
plt.scatter(["setosa","versicolor","virginica"],sgm["SW"])
plt.scatter(["setosa","versicolor","virginica"],sgm["PL"])
plt.scatter(["setosa","versicolor","virginica"],sgm["PW"])
plt.legend(["SL","SW","PL","PW"],fontsize =6)
plt.show()

plt.subplot(223)
plt.plot(["setosa","versicolor","virginica"],sgm["SL"])
plt.plot(["setosa","versicolor","virginica"],sgm["SW"])
plt.plot(["setosa","versicolor","virginica"],sgm["PL"])
plt.plot(["setosa","versicolor","virginica"],sgm["PW"])
plt.legend(["SL","SW","PL","PW"],fontsize =6)
plt.show()

#히스토그램은 x의 값이 의미가 있어야 하나, 꽃 종류에 따른 특징을 나타내는 평균지표에서는 쓰기 어려움
#선그래프는 종류 간 평균 값이 이어져 있어, 그 의미를 명확히 나타내지 못한다.
#점그래프는 그 의미를 이해 할 수는 있지만, x값을 제대로 사용하지 못하고 있어, 조금 산만해 보인다.
#막대그래프는 x좌표의 의미 없이도 종류에 따라 명확히 평균 치를 잘 나타내고 있다.
```

```python
rands = []
rannum = 0

for i in range(0,3):
    for j in range(50*i, 50*i + int(len(irisdf)/12)):
        rannum = np.random.randint(50*i,50+50*i)
        while rannum in rands:
            rannum = np.random.randint(50*i,50+50*i)
        rands.append(rannum)

test_data = irisdf.iloc[rands]
train_data = irisdf.drop(rands,0)

test_data.to_csv("test_dataset.csv")
train_data.to_csv("train_dataset.csv")
```

```python
test_data = pandas.read_csv("test_dataset.csv",sep=',')
train_data = pandas.read_csv("train_dataset.csv",sep=',')

testsg = test_data.groupby("species")
test_sm = testsg.mean()

trainsg = train_data.groupby("species")
train_sm = trainsg.mean()

plt.subplot(221)
plt.bar([i for i in range(0,3)],test_sm["SL"],width=0.1)
plt.bar([i+0.1 for i in range(0,3)],test_sm["SW"],width=0.1)
plt.bar([i+0.2 for i in range(0,3)],test_sm["PL"],width=0.1)
plt.bar([i+0.3 for i in range(0,3)],test_sm["PW"],width=0.1)
plt.xticks([i+0.1 for i in range(0,3)],["setosa","versicolor","virginica"])
plt.title("mean by speicies (test) ")
plt.legend(["SL","SW","PL","PW"],fontsize =7)
plt.show()

plt.subplot(222)
plt.bar([i for i in range(0,3)],train_sm["SL"],width=0.1)
plt.bar([i+0.1 for i in range(0,3)],train_sm["SW"],width=0.1)
plt.bar([i+0.2 for i in range(0,3)],train_sm["PL"],width=0.1)
plt.bar([i+0.3 for i in range(0,3)],train_sm["PW"],width=0.1)
plt.xticks([i+0.1 for i in range(0,3)],["setosa","versicolor","virginica"])
plt.title("mean by speicies (train) ")
plt.legend(["SL","SW","PL","PW"],fontsize =7)
plt.show()

# test집단과 train집단 : 큰 차이는 없지만 약간 다르다
```
