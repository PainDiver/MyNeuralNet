```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

from collections import Counter
import math
import numpy as np
import matplotlib.pyplot as plt

def dot(v,w):
    return sum(v_i*w_i for v_i,w_i in zip(v,w))

def sum_of_squares(v):
    return dot(v,v)

```

```python
num_friends = [100,49,41,40,25,25,25,26,26,18,81,21,54,2,4,82,15,3,4,5,1,53,58,4,3,69,100]

def make_friend_counts_histogram(plt):
    friend_counts = Counter(num_friends)
    xs = range(101)
    ys = [friend_counts[x] for x in xs]
    plt.bar(xs,ys)
    plt.axis([0,101,0,25])
    plt.title("Histogram of Friend Counts")
    plt.xlabel("# of friends")
    plt.ylabel("# of people")
    plt.show()
    
    
make_friend_counts_histogram(plt)
```

```python
num_points = len(num_friends)
largest_value = max(num_friends)
smallest_value = min(num_friends)
sorted_values = sorted(num_friends)
smallest_value = sorted_values[0]
second_smallest_value = sorted_values[1]
second_largest_value = sorted_values[-2]

print(num_points)
print(largest_value)
print(smallest_value)
print(sorted_values)
print(smallest_value)
print(second_smallest_value)
print(second_largest_value)
```

```python
def mean(x):
    return sum(x)/ len(x)

mean(num_friends)

np.mean(num_friends)
```

```python
def median(v):
    n = len(v)
    sorted_v = sorted(v)
    midpoint = n//2 # dividing number as integar
    
    if n%2 == 1:
        return sorted_v[midpoint]
    else:
        lo = midpoint -1
        hi = midpoint
        return (sorted_v[lo]+sorted_v[hi])/2
     
median(num_friends)

np.median(num_friends)
```

```python
def quantile(x,p):
    p_index = int(p*len(x))
    return sorted(x)[p_index]

for i in range(0,100,25):
    print("%.2f Percentage value" %(i*0.01), quantile(num_friends,i*0.01))

np.percentile(num_friends, [i for i in range(0,100,25)])
```

```python
def mode(x):
    x = [int(float(n)) for n in x]
    counts = Counter(x)
    max_count = max(counts.values())
    return [x_i for x_i, count in counts.items() if count == max_count]

mode(num_friends)
```

```python
def data_range(x):
    return max(x)-min(x)

data_range(num_friends)

np.max(num_friends) - np.min(num_friends)
```

```python
def de_mean(x):
    x_bar = mean(x)
    return [x_i - x_bar for x_i in x]

def variance(x):
    n = len(x)
    deviations = de_mean(x)
    return sum_of_squares(deviations)/(n-1)

variance(num_friends)

np.var(num_friends)
```

```python
def standard_deviation(x):
    return math.sqrt(variance(x))

standard_deviation(num_friends)
np.std(num_friends, dtype = np.float64)


def interquartile_range(x):
    return quantile(x,0.75) - quantile(x,0.25)

interquartile_range(num_friends)
```

```python
daily_minutes = [5,1,5,4,83,78,85,4,5,12,35,4,53,21,51,5,53,45,84,83,2,12,42,62,45,2,100]

def covariance(x,y):
    n = len(x)
    return dot(de_mean(x),de_mean(y))/(n-1)

covariance(num_friends,daily_minutes)

np.cov(num_friends,daily_minutes) # [[분산, x공분산], [분산 y공분산]]
```

```python
def correlation(x,y):
    stdev_x = standard_deviation(x)
    stdev_y = standard_deviation(y)
    if stdev_x > 0 and stdev_y >0:
        return covariance(x,y)/stdev_x/stdev_y
    else:
        return 0
    
correlation(num_friends, daily_minutes)

np.corrcoef(num_friends, daily_minutes)

plt.plot(num_friends,daily_minutes, 'ro')
plt.axis([0,max(num_friends)+10,0,max(daily_minutes)+10])
plt.show()
```

```python
outlier = num_friends.index(100)

print(outlier)

num_friends_good = [x for i,x in enumerate(num_friends) if i!= outlier]

daily_minutes_good = [x for i,x in enumerate(daily_minutes) if i!=outlier]

plt.plot(num_friends_good,daily_minutes_good,'ro')
plt.axis([0,max(num_friends_good)+10,0,max(daily_minutes_good)+10])
plt.show()
```

```python
import pandas

data = pandas.read_csv("height-weight.csv",sep=',')
female,male = data.groupby("Gender")

female = female[1]
male = male[1]

maleHeight =male["Height"].tolist()
maleWeight =male["Weight"].tolist()
femaleHeight = female["Height"].tolist()
femaleWeight = female["Weight"].tolist()



#1
plt.axis([50,80,50,280])
plt.scatter(maleHeight,maleWeight,s=0.1,c='b')
plt.scatter(femaleHeight,femaleWeight,s=0.1,c='r')
plt.legend(["Male","Female"])
plt.show()

#2
print(f"Man Average {np.mean(maleHeight)}")
print(f"Man Average {np.mean(maleWeight)}")
print(f"Woman Average {np.mean(femaleHeight)}")
print(f"Woman Average {np.mean(femaleWeight)}")

#3
print(f"Man Median Height {np.median(maleHeight)}")
print(f"Man Median Weight {np.median(maleWeight)}")
print(f"Woman Median Height {np.median(femaleHeight)}")
print(f"Woman Median Weight {np.median(femaleWeight)}")

#4
#하위 25퍼센트에 해당하는 데이터를 가져옴
print(f"Man Quantiled Height {np.percentile(maleHeight,0.25)}")
print(f"Man Quantiled Weight {np.percentile(maleWeight,0.25)}")
print(f"Woman Quantiled Height {np.percentile(femaleHeight,0.25)}")
print(f"Woman Quantiled Weight {np.percentile(femaleWeight,0.25)}")

#5
print(f"Man Mode Height {mode(maleHeight)}")
print(f"Man Mode Weight {mode(maleWeight)}")
print(f"Woman Mode Height {mode(femaleHeight)}")
print(f"Woman Mode Weight {mode(femaleWeight)}")

#6
print(f"Man Data Range Height {np.max(maleHeight)-np.min(maleHeight)}")
print(f"Man Data Range Weight {np.max(maleWeight)-np.min(maleWeight)}")
print(f"Woman Data Range Height {np.max(femaleHeight)-np.min(femaleHeight)}")
print(f"Woman Data Range Weight {np.max(femaleWeight)-np.min(femaleWeight)}")

#7
print(f"Man Variance Height {np.var(maleHeight)}")
print(f"Man Variance Weight {np.var(maleWeight)}")
print(f"Woman Variance Height {np.var(femaleHeight)}")
print(f"Woman Variance Weight {np.var(femaleWeight)}")


#8
print(f"Man Standard Deviation Height {np.std(maleHeight)}")
print(f"Man Standard Deviation Weight {np.std(maleWeight)}")
print(f"Woman Standard Deviation Height {np.std(femaleHeight)}")
print(f"Woman Standard Deviation Weight {np.std(femaleWeight)}")

#9
print(f"Man Covariance Height and Weight {np.cov(maleHeight,maleWeight)}")
print(f"Woman Covariance Height and Weight {np.cov(femaleHeight,femaleWeight)}")


#10
print(f"Man Correlation Height and Weight {np.corrcoef(maleHeight,maleWeight)}")
print(f"Woman Correlation Height and Weight {np.corrcoef(femaleHeight,femaleWeight)}")
```
