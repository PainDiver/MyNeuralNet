```python
import matplotlib.pyplot as plt
import math

def uniform_pdf(x): # 확률밀도함수
    return 1 if x >= 0 and x < 1 else 0

def uniform_cdf(x): # 누적분포함수
    if x < 0:
        return 0
    elif x <1:
        return x
    else:
        return 1

import numpy as np
x = np.arange(-1,2,0.1)

result_array = np.vectorize(uniform_pdf, otypes =[np.float])(x) 
result_array2 = np.vectorize(uniform_cdf, otypes =[np.float])(x) 


plt.axis([-1,2,-1,1.5])
plt.plot(x,result_array)
plt.show()
plt.plot(x,result_array2)
plt.show()

```

```python
def normal_pdf(x,mu=0,sigma=1): # 정규분포
    sqrt_two_pi = math.sqrt(2*math.pi)
    return (math.exp(-(x-mu)**2/2/sigma**2)/(sqrt_two_pi *sigma))

for sigma_value in [1.2,0.5,1]:
    x = np.arange(-6.0,6.0,0.1)
    result_array = np.vectorize(normal_pdf,otypes=[np.float])(x,sigma =sigma_value)
    plt.plot(x,result_array)
    
plt.axis([-6,6,0,1])
plt.show()

def plot_normal_pdfs(plt):
    xs = [x/10.0 for x in range(-50,50)]
    plt.plot(xs,[normal_pdf(x,sigma=1) for x in xs],'-',label="mu =0,sigma=1")
    plt.plot(xs,[normal_pdf(x,sigma=0.5) for x in xs],':',label="mu =0,sigma=0.5")
    plt.plot(xs,[normal_pdf(x,mu=-1) for x in xs],'-',label="mu =-1,sigma=-1")
    plt.legend()
    plt.show()
    
plot_normal_pdfs(plt)
```

```python
# uniform_pdf,uniform_cdf,normal_pdf


# 확률밀도함수
# 10분간격으로 오는 버스가 있다. 내가 5분미만으로 기다릴 확률은?
# x변수에 따른 확률이 같으므로 균등분포

def uniform_pdf(x): # 확률밀도함수 
    return 0.1 if x >= 0 and x <= 10 else 0

x = np.arange(0,11,0.1)
result = np.vectorize(uniform_pdf, otypes =[np.float])(x)
plt.plot(x,result)

def uniform_pdf(x): # 확률밀도함수 
    return 0.1 if x >= 0 and x <= 5 else 0

wait = np.arange(0,5.11,0.1)
waitMinutes = np.vectorize(uniform_pdf, otypes =[np.float])(wait)
plt.plot(wait,waitMinutes)

plt.xticks(range(0,11))
plt.show()


# 누적분포함수
# 누적적으로 올라가는 확률분포. 10분간 기다리면 버스가 오니 무조건 1을 충족하는 확률이 나오기에 x=10에서 1이된다.
def uniform_cdf(x): # 누적분포함수
    if x < 0:
        return 0
    elif x <=10:
        return x/10
    else:
        return 1

x = np.arange(0,11,0.1)
result = np.vectorize(uniform_cdf, otypes =[np.float])(x)
plt.plot(x,result)
plt.xticks(range(0,11))
plt.show()


# 정규분포
# 정규분포는 가운데 (평균)을 기준으로 양옆으로 밸런스가 잡혀있는 종모양의 분포이다.
# 표준정규분포가 아닌 이상 평균, 표준편차도 다르게 나타난다.

def normal_pdf(x,mu=0,sigma=1): 
    sqrt_two_pi = math.sqrt(2*math.pi)
    return (math.exp(-(x-mu)**2/2/sigma**2)/(sqrt_two_pi *sigma))


x= np.arange(-6,6.1,0.1) 
result = np.vectorize(normal_pdf,otypes =[np.float])(x,mu=0,sigma=1.5)
plt.plot(x,result)
plt.axis([-6,6,0,1])
plt.legend(["mu=0, sigma =1.5"])


# (평균 - 1표준편차 <= data <= 평균 - 2표준편차) 의 범위
data = np.arange(-1,2.1,0.1)
result2 = np.vectorize(normal_pdf,otypes=[np.float])(data,mu=0,sigma=1.5)
plt.bar(data,result2,width =0.1)
plt.show()
print(f"확률밀도: {sum(result2*0.1)}")

# 위 정규분포는 표준편차단위를 1보다 크게잡았기 표준정규분포보다 분산이 높다.
# 즉, 원래 그래프상 1에 잡혀야 하는 데이터가 1.5에 잡히므로, 데이터가 분산되어, 확률밀도가 떨어짐

#표준정규분포
# 정규분포의 가운데 값(평균)은 0이고 편차는 1로 맞춰진 종모양 분포를 표준정규분포라고 한다. 

result = np.vectorize(normal_pdf,otypes =[np.float])(x)
plt.plot(x,result)
plt.axis([-6,6,0,1])
plt.legend(["mu=0, sigma =1(normalized)"])

#(평균 - 1표준편차 <= data <= 평균 - 2표준편차) 의 범위
result3 = np.vectorize(normal_pdf,otypes=[np.float])(data,mu=0,sigma=1)
plt.bar(data,result3,width =0.1)
plt.show()
print(f"확률밀도: {sum(result3*0.1)}")
```
